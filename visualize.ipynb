```python
import pandas as pd

# Read the Sheet1 of Excel file into a DataFrame
df = pd.read_excel('merged_all.xlsx', sheet_name='Sheet1')

# In ra số lượng bản ghi trong DataFrame
print(f'Số lượng bản ghi: {len(df)}')
```

```python
# Tạo biểu đồ phân phối của một cột cụ thể
import matplotlib.pyplot as plt

# Lọc dữ liệu để chỉ giữ lại các bản ghi có trường 'Liên quan?' là 'Yes'
df_filtered = df[df['Liên quan?'] == 'Yes']
print(f'Số lượng bản ghi sau khi lọc: {len(df_filtered)}')

# Xoá các bản ghi có trường question_note hoặc Subject rỗng
df_cleaned = df_filtered.dropna(subset=['Subject', 'question_note'])

# In số bản ghi bị xoá
num_deleted = len(df_filtered) - len(df_cleaned)
print(f'Số lượng bản ghi sau khi xoá rỗng: {len(df_cleaned)}')
print(f'Số bản ghi bị xoá: {num_deleted}')
```

```python
import pandas as pd

df = pd.read_excel("cleaned_data.xlsx")

subject_counts = df['Subject'].value_counts()

print(subject_counts)
```

```python
# Vẽ biểu đồ cột thống kê dữ liệu
import matplotlib.pyplot as plt
subject_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Phân phối số lượng câu hỏi theo chủ đề')
plt.xlabel('Chủ đề')
plt.ylabel('Số lượng câu hỏi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Đọc dữ liệu từ file paraphrased_data.xlsx và thống kê số lượng câu hỏi theo từng nhãn bằng biểu đồ tròn
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('augmented_data.xlsx')

subject_counts = df['Subject'].value_counts()
# Vẽ biểu đồ cột thống kê dữ liệu
subject_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Phân phối số lượng câu hỏi theo chủ đề')
plt.xlabel('Chủ đề')
plt.ylabel('Số lượng câu hỏi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Đếm số lượng câu hỏi theo từng nhãn và in ra
print("Số lượng câu hỏi theo từng nhãn:")
for label, count in subject_counts.items():
    print(f"{label}: {count}")
```

```python
# In ra dữ liệu có độ dài nhỏ nhất trong cột 'question_note'
min_length = df['question_note'].str.len().min()
print(f"Độ dài nhỏ nhất trong cột 'question_note': {min_length}")
```
